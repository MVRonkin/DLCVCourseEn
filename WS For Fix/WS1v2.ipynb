## 🧠 Часть 7. Первая нейронная сеть — классификация спиралей

Спирали — классический пример **нелинейно разделимых** данных. В CV такие паттерны встречаются в сегментации и классификации объектов.

Мы используем `nn.Module` — стандартный способ строить модели в PyTorch.

```python
import torch.nn as nn
import torch.nn.functional as F

# Генерация данных (увеличено количество точек для лучшего обучения)
def make_spirals(n=300):
    t = np.linspace(0, 4*np.pi, n)
    noise = 0.15
    X1 = np.stack([t/(2*np.pi)*np.cos(t), t/(2*np.pi)*np.sin(t)], axis=1) + np.random.randn(n, 2)*noise
    X2 = np.stack([t/(2*np.pi)*np.cos(t+np.pi), t/(2*np.pi)*np.sin(t+np.pi)], axis=1) + np.random.randn(n, 2)*noise
    X = np.vstack([X1, X2])
    y = np.array([0]*n + [1]*n)
    return torch.FloatTensor(X), torch.LongTensor(y)

X, y = make_spirals(300)
print(f"Данные: {X.shape}, классы: {y.shape}")

plt.figure(figsize=(6,6))
plt.scatter(X[y==0,0], X[y==0,1], c='blue', label='Класс 0', alpha=0.6)
plt.scatter(X[y==1,0], X[y==1,1], c='red', label='Класс 1', alpha=0.6)
plt.legend(); plt.title("Спиральные данные (нелинейные)"); plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Пример простой сети (MLP)
class SpiralNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(2, 64)   # вход 2D → скрытый слой
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 2)   # 2 класса
    
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Создаём модель, loss и оптимизатор
model = SpiralNet()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# Обучение (студент может экспериментировать)
for epoch in range(800):
    optimizer.zero_grad()
    outputs = model(X)
    loss = criterion(outputs, y)
    loss.backward()
    optimizer.step()
    
    if epoch % 200 == 0:
        print(f"Epoch {epoch:3d} | Loss: {loss.item():.4f}")

# Оценка
with torch.no_grad():
    preds = model(X).argmax(dim=1)
    acc = (preds == y).float().mean().item() * 100
print(f"\n🎉 Точность на обучающих данных: {acc:.1f}% (цель — > 90%)")
```
