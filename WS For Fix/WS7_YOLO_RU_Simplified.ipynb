- **Очень быстрая** (реал-тайм даже на CPU)
- **Простая** в использовании
- **Высокая точность** (особенно YOLOv8, YOLOv11)
- Отлично подходит для мобильных и embedded устройств
- Активно развивается (Ultralytics)

# 2. Быстрый старт: Предобученная модель

```python
# Загружаем предобученную YOLOv8n (nano)
model = YOLO('yolov8n.pt')

# Инференс на изображении
results = model('https://ultralytics.com/images/bus.jpg')
results[0].show()  # Показать результат
```

# 3. Обучение на своём датасете

YOLO использует формат **YOLO** (один .txt файл на изображение с координатами bounding boxes).

```python
# Пример структуры датасета:
# dataset/
#   images/
#     train/
#     val/
#   labels/
#     train/
#     val/
#   data.yaml
```

```python
# data.yaml пример:
# train: ./images/train
# val: ./images/val
# nc: 2
# names: ['cat', 'dog']
```

```python
# Обучение
model = YOLO('yolov8n.pt')

results = model.train(
    data='data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    device=0 if torch.cuda.is_available() else 'cpu',
    project='runs/detect',
    name='custom_yolo'
)
```

# 4. Метрики оценки

```python
# После обучения можно посмотреть метрики
metrics = model.val()
print(f"mAP50-95: {metrics.box.map:.3f}")
print(f"mAP50:    {metrics.box.map50:.3f}")
print(f"Precision: {metrics.box.mp:.3f}")
print(f"Recall:    {metrics.box.mr:.3f}")
```

# 5. Инференс и визуализация

```python
# Инференс на видео или папке
results = model.predict(
    source='path/to/images_or_video',
    conf=0.25,      # порог уверенности
    iou=0.45,       # порог NMS
    save=True       # сохранить результаты
)
```
